```python
%%time
import re
import os
import shutil
import zipfile
from zipfile import ZipFile
from time import sleep
import rarfile
import patoolib
import pandas as pd
```

```python
def files_create_folders(form, file, names):
    error = []
    temp = str(form)
    zipfile = file
    BASE_DIR = os.path.abspath(r'C:\Users\DSimonov\Desktop\forms')
#    print("Setting default working folder")
    os.chdir(BASE_DIR)
#    print("Changing rights for working folder")
    os.chmod (BASE_DIR, 0o777)
    temp_path = os.path.join(BASE_DIR, temp)
#    shutil.rmtree(temp_path)
    try:
        if not os.path.exists(temp_path):
            os.makedirs(temp_path)
            os.chmod (temp_path, 0o777)
    except OSError:
        print ('Error: Creating directory. ' +  temp_path)
#    print(f"taking file {file}")
#    print(f"from the path {BASE_DIR}")
#    print(f"to the path {temp_path}")
    shutil.copy(file, temp_path)
#    sleep(5)
#    print(file)
    os.chdir(temp_path)
    if file.endswith(".zip"):    
        unarchive_zip(temp_path, file, errors)
    if file.endswith(".rar"):
        unarchive_rar(temp_path, file, errors)
#    print(f"Now file name is {file}")
#    os.remove(file)
#    print("changing default working folder")
    os.chdir(temp_path)
#    print("creating new temporary working folder")
    temp_path_zip = os.path.join(temp_path, "ready_for")
    try:
        if not os.path.exists(temp_path_zip):
            os.makedirs(temp_path_zip)
    except OSError:
        print ('Error: Creating directory. ' +  temp_path_zip  )  
    for file in os.listdir(temp_path):
#        os.remove(zipfile)
        if file.endswith("B1.DBF") or file.endswith("_P1.DBF") or file.endswith("123D.DBF") or file.endswith("134D.DBF") or file.endswith("135_3.dbf"):
            print(f"found file to move: {file} and moving to {temp_path_zip}")
            shutil.move(file, temp_path_zip)
        elif file.endswith("for"):
            pass
        else:
            os.remove(file)          
#    print("changing default working folder")
    os.chdir(BASE_DIR)
#    print("finished cycle!")
    if error:
        errors.append(error)
        print(f'appending error: {errors}, len is {len(errors)}')
    return
```

```python
def unarchive_zip(temp_path, file, errors):
    os.chmod (temp_path, 0o777)
    temp_path_zip = os.path.join(temp_path, "ready_for")
    try:
        if not os.path.exists(temp_path_zip):
            os.makedirs(temp_path_zip)
    except OSError:
        print ('Error: Creating directory. ' +  temp_path_zip)
    try:
        with ZipFile(file, 'r') as f:
            f.extractall()
            print('Unzip Done!')
    except:
        errors.append(file)
        print(f"Unzip error!!! len = {len(file)}")
    return
```

```python
def unarchive_rar(temp_path, file, errors):
    os.chmod (temp_path, 0o777)
    temp_path_zip = os.path.join(temp_path, "ready_for")
    try:
        if not os.path.exists(temp_path_zip):
            os.makedirs(temp_path_zip)
    except OSError:
        print ('Error: Creating directory. ' +  temp_path_zip)
    try:
        patoolib.extract_archive(file, outdir=temp_path)
        print('Unrar Done!')
    except:
        print(f"Unrar error!!! len = {len(file)}")
        errors.append(file)
    return
```

```python
%%time
print(f"Starting phase 1: unpacking archives from CBR site")
BASE_DIR = os.path.abspath(r'C:\Users\DSimonov\Desktop\forms')
os.chdir(BASE_DIR)
files = [f for f in os.listdir(r'C:\Users\DSimonov\Desktop\forms') if os.path.isfile(f)]
print(f"Found {len(files)} files")
errors = []
for file in files:
    if re.match('(\d\d\d)',file):
        forms = re.match('(\d\d\d)',file).group()
        names = re.findall('-(\d+).', file)
        files_create_folders(forms, file, names)
print(f'Found {len(errors)} errors')
df_errors = pd.DataFrame(errors)
print(f'Saving file with errors')
df_errors.to_csv('errors.csv')
print("Completed phase 1")
```
